# Predicting Income with Random Forests
**Objective:** Use a random forest on a census dataset to predict whether a chosen person makes more than $50,000.

**Question:** What variables do you think might be helpful in predicting income?\
**Answer:** I think the following can help make such predictions:
* ```workclass```

* ```capital-loss```
* ```hours-per-week```

````python
# 2. Load the income dataset.
income_data = pd.read_csv('income.csv', header=0)
````

````python
# 3. Look at the first row.
print('Data on the first person in this dataset:')
print(income_data.iloc[0])
````

**Question:** Did this person make more than $50,000?\
**Answer:** No, he hadn't.

**Question:** What is the name of the column that contains that information?\
**Answer:** The column is named ```income```.

````python
# 4. Eliminate all extra spaces.
income_data = pd.read_csv('income.csv', header=0, delimiter=',')
````

## Format the Data for Scikit-Learn

````python
# 5. Glean the incomes (henceforth known as labels).
income_data.columns = income_data.columns.str.strip()
labels = income_data[['income']]
````

````python
# 6. Glean the columns that best predict income (henceforth known as data).
data = income_data[['age', 'capital-gain', 'capital-loss', 'hours-per-week', 'sex']]
````

````python
# 7. Split the DataFrames above into training and testing sets.
train_data, test_data, train_labels, test_labels = train_test_split(data, labels, random_state=1)
````

## Create the Random Forest

````python
# 8. Create a RandomForestClassifier object.
forest = RandomForestClassifier(random_state=1)
````

````python
# 9. Fit the forest to the training set (this will cause an error).
forest.fit(train_data, train_labels)
````

````python
# 10. Remove the `sex` column from data and try fitting the dataset again.
data = data.drop(columns='sex')
train_data, test_data, train_labels, test_labels = train_test_split(data, labels, random_state=1)
forest.fit(train_data, train_labels)
````

````python
# 11. Test the forest's accuracy.
print('Accuracy (%) without `sex`:', forest.score(test_data, test_labels)*100.0)
````

## Changing Column Types

````python
# 12-13a. Introduce a new column that maps one's sex to a binary number: 0 for males and 1 for females.
income_data['sex-int'] = income_data['sex'].apply(lambda row: 0 if row=='Male' else 1)
data = income_data[['age', 'capital-gain', 'capital-loss', 'hours-per-week', 'sex-int']]
````

````python
# 13b. Test the new forest's accuracy. 
train_data, test_data, train_labels, test_labels = train_test_split(data, labels, random_state=1)
forest.fit(train_data, train_labels)
print('Accuracy (%) with `sex-int`:', forest.score(test_data, test_labels)*100.0)
````

````python
# 14. Print how many times each country appears in the dataset.
print('Here\'s how often each country in our dataset appeared:')
print(income_data['native-country'].value_counts())
````

````python
# 15-16a. Introduce a new column with the following binary values -- 0 if you're from the United States, and 1 if you're not.
income_data['country-int'] = income_data['native-country'].apply(lambda row: 0 if row=='United-States' else 1)
data = income_data[['age', 'capital-gain', 'capital-loss', 'hours-per-week', 'sex-int', 'country-int']]
train_data, test_data, train_labels, test_labels = train_test_split(data, labels, random_state=1)
forest.fit(train_data, train_labels)
````

````python
# 16b. Determine how `country-int` changes the model's accuracy.
print('Accuracy (%) with `country-int`:', forest.score(test_data, test_labels)*100.0)
````
