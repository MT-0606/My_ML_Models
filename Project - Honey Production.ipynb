# Objective: Investigate the decline in honeybee populations using linear regression.

````python
# Libraries
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn import linear_model
````

````python
# Dataset
df = pd.read_csv('honeyproduction.csv')
````

## Check out the Data

### Task 1: Get a sense of how ```df``` is constructed.

````python
print('Here\'s a preview of the honey production dataframe:')
print(df.head())
````

### Task 2: Get the mean of ```totalprod``` (how much honey was produced in all 50 states) per year.

````python
prod_per_year = df.groupby('year')['totalprod'].mean().reset_index()
````

### Tasks 3-5: Produce the scatter plot between ```X``` (years) and ```y``` (total honey produced).

````python
X = prod_per_year['year']
X = X.values.reshape(-1, 1)
````

````python
y = prod_per_year['totalprod']
````

````python
plt.scatter(X, y)
plt.show()
````

## Create and Fit a Linear Regression Model

### Task 6: Create the model.

````python
regr = linear_model.LinearRegression()
````

### Task 7-8: Fit the model to the table, then obtain the slope and intercept of the line of best fit.

````python
regr.fit(X, y)
print('Slope:', regr.coef_[0])
print('Intercept:', regr.intercept_)
````

### Tasks 9-10: Get and plot all predictions from the ```regr``` model.

````python
y_predict = regr.predict(X)
plt.scatter(X, y)
plt.plot(X, y_predict, color='red')
plt.xlabel('Year')
plt.ylabel('Average Amount of Honey Produced')
plt.show()
````

## Predict the Honey Decline

### Tasks 11-13: Predict how much honey will be produced in 2050.

````python
X_future = np.array(range(2013, 2050))
X_future = X_future.reshape(-1, 1)
````

````python
future_predict = regr.predict(X_future)
````

````python
plt.plot(X_future, future_predict)
plt.show()
````

**Verdict:** If honey production keeps dropping at this rate, 0.40 million gallons of honey will be produced in 2050.
